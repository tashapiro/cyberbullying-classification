## Create New Features

```python
#https://stackoverflow.com/questions/11331982/how-to-remove-any-url-within-a-string-in-python
def remove_urls (vTEXT):
    vTEXT = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', vTEXT, flags=re.MULTILINE)
    return(vTEXT)
```

```python
#remove links 
df['no_links_text'] = [remove_urls(i) for i in df['tweet_text']]

#character and word length
df['char_len'] = [len(i) for i in df.no_links_text]
df['word_count'] = [len(i.split()) for i in df.no_links_text]


df['contains_link'] = np.where(df['tweet_text'].str.contains('http'),True,False)
df['is_retweet'] = np.where(df['tweet_text'].str.contains('RT'),True,False)


df.sample(5)
```

## Functions to Facilitate EDA

```python
def remove_urls (vTEXT):
    vTEXT = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', vTEXT, flags=re.MULTILINE)
    return(vTEXT)

df['no_links_text'] = [remove_urls(i) for i in df['tweet_text']]

def get_top_text(stop = 'english', ngram = (1,1), class_type = 'religion', top_n = 25):
    #intialize cv
    cv = CountVectorizer(stop_words='english',ngram_range = ngram)
    #subset df
    subset = df[df['cyberbullying_type']==class_type]
    cv.fit(subset['no_links_text'])
    cv_text = cv.transform(subset['no_links_text'])
    new_cv = pd.DataFrame(cv_text.todense(),columns=cv.get_feature_names())
    top_text = pd.DataFrame(new_cv.sum().T).sort_values(by=0, ascending=False).head(top_n)
    top_text.rename(columns={0:'Count'},inplace=True)
    return top_text
```

```python
religion_10 = get_top_text('english',(2,2),'religion',10)
gender_10 = get_top_text('english',(2,2),'gender',10)
age_10 = get_top_text('english',(2,2),'age',10)
ethnicity_10 = get_top_text('english',(2,2),'ethnicity',10)
other_10 = get_top_text('english',(2,2),'other_cyberbullying',10)
```

## Single Word Visual

```python
religion_10 = get_top_text('english',(1,1),'religion',10)
gender_10 = get_top_text('english',(1,1),'gender',10)
age_10 = get_top_text('english',(1,1),'age',10)
ethnicity_10 = get_top_text('english',(1,1),'ethnicity',10)
other_10 = get_top_text('english',(1,1),'other_cyberbullying',10)
```

```python
f, axes = plt.subplots(2, 2)

color = '#E75A7C'

#sns.set_style("whitegrid")
sns.set(rc = {'figure.figsize':(15,10)})

(sns.barplot(x="Count",  
            y= "Bigram", 
            color='#008BF8',
            ax=axes[0][0],
            data=religion_10.reset_index().rename(columns={"index":"Bigram"}))
 .set(ylabel=None, title="Religion")
)

(sns.barplot(x="Count",  
            y= "Bigram", 
            color='#DC0073',
            ax=axes[1][0],
            data=gender_10.reset_index().rename(columns={"index":"Bigram"}))
 .set(ylabel=None, title="Gender")
)


(sns.barplot(x="Count",  
            y= "Bigram", 
            color='#F5B700',
            ax=axes[0][1],
            data=age_10.reset_index().rename(columns={"index":"Bigram"}))
 .set(ylabel=None, title="Age")
)


(sns.barplot(x="Count",  
            y= "Bigram", 
            color='#5F4BB6',
            ax=axes[1][1],
            data=ethnicity_10.reset_index().rename(columns={"index":"Bigram"}))
 .set(ylabel=None, title="Ethnicity")
)

f.suptitle("Top 10 Words by Cyberbullying Classification")

plt.savefig('../plots/top_words.png')
```

## Character & Word Count Distribution

```python
char_len = (ggplot(df, aes(x='char_len', fill="cyberbullying_type"))
                +geom_histogram(bins=30) 
                +facet_wrap("cyberbullying_type")
                +scale_x_continuous(limits=(0,300))
                + ylab("Count")
                + xlab("Character Length")
                +ggtitle("Character Length Distribution")
                +theme(figure_size=(12, 4),
                    text = element_text(family="Roboto"),
                    plot_title=element_text(weight='bold',color='black', size=14),
                    panel_background = element_rect(fill="white"),
                    panel_grid_major = element_line(color="#dee2e6"),
                    subplots_adjust={'wspace': 0.25},
                    legend_position="none",
                    axis_title = element_text(weight='bold'))
          )

ggsave(plot=char_len, filename='../plots/char_len_dist.png', dpi=1000)

char_len
```

```python
#set up group by averages per class, this will help us reorder the boxplots by descending avg
class_char_avg = pd.DataFrame(df.groupby('cyberbullying_type')['char_len'].mean()).rename(columns={'char_len':'avg_char'}).reset_index()
char_df = df.merge(class_char_avg, on='cyberbullying_type')
```

```python
 box_char_len = (ggplot(char_df, aes(x= "reorder(cyberbullying_type,avg_char)", y= "char_len"))
    + geom_boxplot(fill="#26C485")
    + xlab("Classification")
    + ylab("Character Length")
    + ggtitle("CHARACTER LENGTH")
    + scale_y_continuous(limits=(0,400))
    + geom_hline(aes(yintercept = df.char_len.mean()), linetype='dotted',size = 0.8) 
    + coord_flip()
    +theme(figure_size=(10, 6),
            text = element_text(family="Roboto"),
            plot_title=element_text(weight='bold',color='black', size=12),
            legend_position=None,
            panel_background = element_rect(fill="white"),
            panel_grid_major = element_line(color="#dee2e6"),
            axis_title = element_text(weight='bold'))
 )
    
ggsave(plot=box_char_len, filename='../plots/box_char_len.png', dpi=1000)

box_char_len
```

```python
word_dist = (ggplot(df, aes(x='word_count', fill="cyberbullying_type"))
                +geom_histogram(bins=30) 
                +facet_wrap("cyberbullying_type")
                +scale_x_continuous(limits=(0,75))
                + ylab("Count")
                + xlab("Word Count")
                +ggtitle("Word Count")
                +theme(figure_size=(12, 4),
                    text = element_text(family="Roboto"),
                    plot_title=element_text(weight='bold',color='black', size=14),
                    panel_background = element_rect(fill="white"),
                    panel_grid_major = element_line(color="#dee2e6"),
                    subplots_adjust={'wspace': 0.25},
                    legend_position="none",
                    axis_title = element_text(weight='bold'))
          )

ggsave(plot=word_dist, filename='../plots/word_count_dist.png', dpi=1000)

word_dist
```

```python
#set up group by averages per class, this will help us reorder the boxplots by descending avg
class_word_avg = pd.DataFrame(df.groupby('cyberbullying_type')['word_count'].mean()).rename(columns={'word_count':'avg_words'}).reset_index()
word_df = df.merge(class_word_avg, on='cyberbullying_type')
```

```python
box_word_count = (ggplot(word_df, aes(x= "reorder(cyberbullying_type, avg_words)", y= "word_count"))
    + geom_boxplot(fill="#26C485")
    + xlab("Classification")
    + ylab("Word Count")
    + ggtitle("WORD COUNT")
    + scale_y_continuous(limits=(0,75))
    + geom_hline(aes(yintercept = df.word_count.mean()), linetype='dotted',size = 0.8) 
    + coord_flip()
    +theme(figure_size=(10, 6),
            text = element_text(family="Roboto"),
            plot_title=element_text(weight='bold',color='black', size=14),
            legend_position=None,
            panel_background = element_rect(fill="white"),
            panel_grid_major = element_line(color="#dee2e6"),
            axis_title = element_text(weight='bold'))
 )


ggsave(plot=box_word_count, filename='../plots/box_word_count.png', dpi=1000)

box_word_count
```
